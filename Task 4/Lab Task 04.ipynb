 Implement the N-Queens Problem (Dynamic)

```python

def is_safe(board, row, col, n):   
 
 for i in range(row):
  if board[i] == col:
   return False    
 i, j = row - 1, col - 1
 while i >= 0 and j >= 0:
  if board[i] == j:
   return False
  i -= 1
  j -= 1
    
 i, j = row - 1, col + 1
 while i >= 0 and j < n:
  if board[i] == j:
   return False
  i -= 1
  j += 1
    
 return True

def print_board(board, n):
 print("\n" + "="*(2*n+1))
 print("CHESSBOARD SOLUTION:")
 print("="*(2*n+1))
    
 for i in range(n):
  print("|", end="")
  for j in range(n):
   if board[i] == j:
    print("Q|", end="")
   else:
    print(" |", end="")
  print()
    
 print("="*(2*n+1))

def print_coordinates(board, n):
 print("\nQueen Positions (row, column):")
 print("-"*30)
 for i in range(n):
  print(f"Queen {i+1}: ({i}, {board[i]})")

def solve_n_queens_util(board, row, n, solutions, find_all=False):
 if row == n:
  solutions.append(board.copy())
  return True if not find_all else False
    
 for col in range(n):
  if is_safe(board, row, col, n):            
   board[row] = col
   print(f"Step: Placed queen at row {row}, column {col}")
   if find_all:
    solve_n_queens_util(board, row + 1, n, solutions, find_all)
   else:
    if solve_n_queens_util(board, row + 1, n, solutions, find_all):
     return True
            
   board[row] = -1
   print(f"Step: Backtracked from row {row}")
    
 return False

def solve_n_queens(n, find_all=False):
 board = [-1] * n    
 solutions = []
 print("\n" + "="*60)
 print(f"SOLVING {n}-QUEENS PROBLEM USING BACKTRACKING")
 print("="*60)
 print(f"Board Size: {n} x {n}")
 print(f"Mode: {'Find ALL solutions' if find_all else 'Find ONE solution'}")
 print("-"*60)
 if find_all:
  solve_n_queens_util(board, 0, n, solutions, find_all)
 else:
  solve_n_queens_util(board, 0, n, solutions, find_all)
   
 return solutions

def main():
 print("="*60)
 print("N-QUEENS Peoblem Solver (Backtracking only)")
 print("="*60)
 try:
  n = int(input("\nEnter number of queens (N): "))
  if n <= 0:
   print("Please enter a positive integer!")
   return
  if n < 4 and n > 1:
   print(f"Note: {n}-Queens problem has no solution!")
 except ValueError:
  print("Please enter a valid integer!")
  return
 print("\nChoose option:")
 print("1. Find ONE solution")
 print("2. Find ALL solutions")
    
 try:
  choice = int(input("Enter choice (1 or 2): "))
 except ValueError:
  print("Invalid choice! Using option 1 (ONE solution)")
  choice = 1
    
 find_all = (choice == 2)    
 import time
 start_time = time.time()
 solutions = solve_n_queens(n, find_all)
 end_time = time.time()
 print("\n" + "="*60)
 print("RESULTS")
 print("="*60)  
 if solutions:
  if find_all:
   print(f"Total solutions found: {len(solutions)}")
   print(f"Time taken: {end_time - start_time:.4f} seconds")
   if len(solutions) > 0:
    see_all = input("\nDo you want to see all solutions? (y/n): ").lower()
    if see_all == 'y':
     for idx, solution in enumerate(solutions, 1):
      print(f"\nSolution {idx}:")
      print_board(solution, n)
      print_coordinates(solution, n)
      if idx < len(solutions):
       input("\nPress Enter to see next solution...")
   if len(solutions) > 0:
    print("\n" + "="*60)
    print("First Solution (Detailed View)")
    print("="*60)
    print_board(solutions[0], n)
    print_coordinates(solutions[0], n)
   else:
    print("Solution found!")
    print(f"Time taken: {end_time - start_time:.4f} seconds")
    print_board(solutions[0], n)
    print_coordinates(solutions[0], n)
 else:
  print(f"No solution exists for {n}-Queens problem!")
  print(f"Time taken: {end_time - start_time:.4f} seconds")
    
 print("\n" + "="*60)
 print("Program completed")
 print("="*60)
main()
```
