 Implement the Water Jug Problem using Depth-First Search (DFS) and print the rules applied at each step. 

```python
Cap_A = 4
Cap_B = 3
Goal = 2

visited_states = set()

def dfs(current_state, steps):
 x, y = current_state    
 if x == Goal or y == Goal:
  print("\nGoal Achieved!\n")
  for s in steps:
   print(s)
  return True
    
 visited_states.add(current_state)    
 possible_actions = []
    
 # 1. Fill Jug A 
 if x < Cap_A:
  possible_actions.append(((Cap_A, y), "Fill Jug A"))
    
 # 2. Fill Jug B 
 if y < Cap_B:
  possible_actions.append(((x, Cap_B), "Fill Jug B"))
      
 # 3. Empty Jug A
 if x > 0:
  possible_actions.append(((0, y), "Empty Jug A"))
      
 # 4. Empty Jug B 
 if y > 0:
  possible_actions.append(((x, 0), "Empty Jug B"))
    
 # 5. Pour Jug A into Jug B 
 if x > 0 and y < Cap_B:
  transfer = min(x, Cap_B - y)
  possible_actions.append(((x - transfer, y + transfer), "Pour Jug A into Jug B"))
    
 # 6. Pour Jug B into Jug A 
 if y > 0 and x < Cap_A:
  transfer = min(y, Cap_A - x)
  possible_actions.append(((x + transfer, y - transfer), "Pour Jug B into Jug A"))
    
 for next_state, rule in possible_actions:
  if next_state not in visited_states:
   new_steps = steps + [f"{current_state} → {next_state}  |  {rule}"]
   if dfs(next_state, new_steps):
    return True
    
 return False

initial_state = (0, 0)
dfs(initial_state, [])

```
